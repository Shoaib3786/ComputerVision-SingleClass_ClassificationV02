## Importing Dataset

```python
# get this Image file [in sequence]
xdata = ["/newDataset/Frame({}).jpg".format(i) for i in range(0, 101)]

# get this corrdinates and class file [in sequence]
ydata = ["/newDataset/Frame({}).txt".format(i) for i in range(0, 101)]

# check if above exist then import then get the file in list
fileX = [f for f in xdata if isfile(f)]
print(len(fileX))

# check if above exist then get the file in list
fileY = [f for f in ydata if isfile(f)]
print(len(fileY))



### now getting the content of the .txt file in csv so that we can convert in pd dataframes
for x in fileY:          
    with open(x) as fin, open('/newDataset/y_train.csv', 'a') as fout:
        y_csv_output=csv.writer(fout)
        for line in fin:
            y_csv_output.writerow(line.split())


### now getting the content of the fileX in csv format so that we can merge them in pd dataframes
with open('/newDataset/x_train.csv','w') as f:
    for x in fileX:
        f.write(str(x))
        f.write('\n')
            
```

```python
### Pandas work to get y_train.csv [consist of class + coordinates] ###
y_data = pd.read_csv('/content/y_train.csv', names= ['class', 'xmin', 'xmax', 'ymin', 'ymax'])
y_df = pd.DataFrame(data=y_data) 
y_df.head()

### Pandas work to get x_train.csv [consist images] ###
x_data = pd.read_csv('/content/x_train.csv', names= ['path'])
x_df = pd.DataFrame(data = x_data) 
x_df.head()


#combine dataframe [consist of path + coordinates + class]
df = pd.concat([x_df, y_df], axis="columns")
print(df)
```

```python

# making list for fileX name it as x_df [consist of path of image]
img_path_train = df['path'].tolist()
print('img-',len(x_df))

# making array of class using y_df [consist of only class]
class_train = np.array(y_df['class'])
print('class-',len(class_train))  # for verifiction


# making array of regression using y_df [consist of only coordinate]
col = ['xmin', 'xmax', 'ymin', 'ymax']
reg_train = np.array(y_df[col])
print('col-',len(reg_train))   # for verifiction

```

#### Converting above x_df, y_df into tensorflow dataset:
 - Using tf_dataset_slices
 - Saving the images in tensor format not in text format, hence have to use load_images() to decode the image file path.

```python

train_dataset = tf.data.Dataset.from_tensor_slices((img_path_train, 
                                                {"class":class_train, "regressor":reg_train}) )

print(len(train_dataset))

# # from_tensor_slices is used to build tensorflow dtype dataset
# test_dataset = tf.data.Dataset.from_tensor_slices((test_df['path'].tolist(), 
#                                                     {"class":test_class_arr, "regressor":test_bbox_arr}) )


# process of getting the image through the path and return image as matrix & lable dict of {class, regressor} as it is
def load_image(path, label):

  # read the file from the path
  image = tf.io.read_file(path)

  #now to decode the file from previous readfile because it read the file and stored it in Machine language now we have to decode it 
  image = tf.image.decode_png(image, channels=1)
  
  # # convert the channels from 1 to 3 bcz model accepts 3 channel
  image = tf.image.grayscale_to_rgb(image)

  image = tf.cast(image, dtype=tf.float32) / 255. #after typecast do normalize also

  # image = tf.expand_dims(image, axis=0)
  # image = tf.expand_dims(image, axis=0)
  return image, label


# Now to successfully replace the train_df path containing part with actuall image matrix as well as with dictonary
# of class and regresso value, all this inside tensorflow dataset.

# using map() with tensorflow dataset [tf_Dataset.map(func)]  --> tf_dataset is like iterable ---> see documentation
train_dataset = train_dataset.map(load_image).batch(2)  #randomly assigned batch size, 

# test_dataset = test_dataset.map(load_image).batch(4)   #randomly assigned batch size


len(train_dataset)  # now len will decrese bcz of batch 4,  229/4
```
